```python
# enriched_airlinepassenger.ipynb

import pandas as pd
import sqlite3

# Load datasets
passenger_data = pd.read_csv('airline_passenger_satisfaction.csv')
mock_data = pd.read_csv('MOCK_DATA.csv')

# Display column names to verify
print("Columns in passenger_data:")
print(passenger_data.columns)
print("\nColumns in mock_data:")
print(mock_data.columns)

# Rename columns to match for merging
passenger_data.rename(columns={'ID': 'Passenger ID'}, inplace=True)
mock_data.rename(columns={'Passenger ID': 'PassengerId'}, inplace=True)

# Clean column names by stripping whitespace (if needed)
passenger_data.columns = passenger_data.columns.str.strip()
mock_data.columns = mock_data.columns.str.strip()

# Check for 'PassengerId' column after renaming
if 'Passenger ID' in passenger_data.columns and 'PassengerId' in mock_data.columns:
    # Merge datasets
    merged_data = pd.merge(passenger_data, mock_data, left_on='Passenger ID', right_on='PassengerId')
    
    # Save merged data to a CSV file
    merged_data.to_csv('merged_data.csv', index=False)
    
    # Create a connection to SQLite database and save merged data to a table
    conn = sqlite3.connect('airline_insights.db')
    merged_data.to_sql('merged_data', conn, if_exists='replace', index=False)
    
    # Display the first few rows of the merged data
    print("\nMerged Data:")
    print(merged_data.head())
    
    # Close the database connection
    conn.close()
else:
    print("Error: 'Passenger ID' column not found in one of the datasets.")





```
